```python
def WartOA(obj_func, dim, N, max_iter, lb, ub):
    if not isinstance(lb, np.ndarray): lb = np.full(dim, lb)
    if not isinstance(ub, np.ndarray): ub = np.full(dim, ub)
    pop = lb + (ub - lb) * np.random.rand(N, dim)
    fitness = np.array([obj_func(ind, dim) for ind in pop])
    prev_locations = np.copy(pop)
    sw_idx = np.argmin(fitness)
    sw_pos = pop[sw_idx]
    sw_fitness = fitness[sw_idx]
    ages = np.random.randint(1, 16, N)
    time_cycle = ['morning', 'mid-day', 'late-afternoon', 'night']
    weather_conditions = ['hot', 'rainy', 'normal']
    time_idx = 0
    num_burrows = random.randint(15, 25)
    burrows = lb + (ub - lb) * np.random.rand(num_burrows, dim)
    convergence_history = []
    for t in range(max_iter):
        current_best_idx = np.argmin(fitness)
        if fitness[current_best_idx] < sw_fitness:
            sw_pos = pop[current_best_idx]
            sw_fitness = fitness[current_best_idx]
        if (t + 1) % 10 == 0:
            num_burrows = random.randint(15, 25)
            burrows = lb + (ub - lb) * np.random.rand(num_burrows, dim)
        current_time = time_cycle[time_idx % len(time_cycle)]
        weather = random.choice(weather_conditions)
        time_idx += 1
        for i in range(N):
            group_i = random.uniform(1, 2) if ages[i] > 2 and np.random.rand() < 0.5 else random.uniform(2, 3)
            new_pos = np.copy(pop[i])
            if current_time in ['morning', 'late-afternoon'] and weather == 'normal':
                feeding_activity = np.random.rand()
                if current_time == 'late-afternoon':
                    a = random.randint(1, 3)
                    new_pos = pop[i] + feeding_activity * a * (sw_pos - group_i * pop[i])
                else:
                    new_pos = pop[i] + feeding_activity * (sw_pos - group_i * pop[i])
                    if t >= 4 and obj_func(new_pos, dim) > obj_func(prev_locations[i], dim):
                        new_pos = prev_locations[i]
            elif current_time == 'mid-day' or weather in ['hot', 'rainy']:
                resting_activity = np.random.rand()
                shade = lb + (ub - lb) * np.random.rand(dim)
                new_pos = pop[i] + resting_activity * (shade - pop[i])
            elif current_time == 'night':
                prev_locations[i] = np.copy(pop[i])
                distances = np.linalg.norm(pop[i] - burrows, axis=1)
                closest_burrow = burrows[np.argmin(distances)]
                new_pos = np.random.rand() * closest_burrow
            new_pos = np.clip(new_pos, lb, ub)
            new_fitness = obj_func(new_pos, dim)
            if new_fitness < fitness[i]:
                pop[i], fitness[i] = new_pos, new_fitness
        if (t + 1) % 20 == 0:
            ages = np.random.randint(1, 16, N)
        convergence_history.append(sw_fitness)
    return sw_pos, sw_fitness, convergence_history
```

```python
def plot_convergence_with_forced_ranges(history, function_name):
    """
    Plots the convergence history, forcing the Y-axis ranges to match
    the visualization in the GRO paper for better comparison.
    """
    plt.figure(figsize=(10, 6))

    # Define the scale and Y-limits for each function based on the GRO paper
    # (These are approximations based on Figures 12, 13, and 14)
    plot_config = {
        'f1':  {'scale': 'log', 'ylim': (1e-300, 1e100)},
        'f2':  {'scale': 'log', 'ylim': (1e-200, 1e50)},
        'f3':  {'scale': 'log', 'ylim': (1e-100, 1e5)},
        'f4':  {'scale': 'log', 'ylim': (1e-200, 1e50)},
        'f5':  {'scale': 'log', 'ylim': (1e1, 1e15)},
        'f6':  {'scale': 'log', 'ylim': (1e-10, 1e5)},
        'f7':  {'scale': 'log', 'ylim': (1e-10, 1e2)},
        'f8':  {'scale': 'linear', 'ylim': (-14000, -2000)},
        'f9':  {'scale': 'linear', 'ylim': (0, 500)},
        'f10': {'scale': 'log', 'ylim': (1e-15, 1e2)},
        'f11': {'scale': 'log', 'ylim': (1e-15, 1e5)},
        'f12': {'scale': 'log', 'ylim': (1e-15, 1e10)},
        'f13': {'scale': 'log', 'ylim': (1e-15, 1e10)},
        'f14': {'scale': 'log', 'ylim': (1e0, 1e2)},
        'f15': {'scale': 'log', 'ylim': (1e-5, 1e2)},
        'f16': {'scale': 'linear', 'ylim': (-1.1, 5)},
        'f17': {'scale': 'linear', 'ylim': (0.39, 0.43)},
        'f18': {'scale': 'linear', 'ylim': (2.9, 5)},
        'f19': {'scale': 'linear', 'ylim': (-3.9, -3.75)},
        'f20': {'scale': 'linear', 'ylim': (-3.5, -2.5)},
        'f21': {'scale': 'log', 'ylim': (1e1, -1e1)}, # Linear in paper, but data is negative
        'f22': {'scale': 'log', 'ylim': (1e1, -1e1)}, # Linear in paper, but data is negative
        'f23': {'scale': 'log', 'ylim': (1e1, -1e1)}, # Linear in paper, but data is negative
        # Default for any functions not listed
        'default': {'scale': 'linear', 'ylim': None}
    }

    config = plot_config.get(function_name.lower(), plot_config['default'])

    # Handle zero values for log scale by replacing them with a very small number
    if config['scale'] == 'log':
        plt.yscale('log')
        history_plot = [val if val > 0 else 1e-300 for val in history] # Use a very small epsilon
    else:
        history_plot = history

    plt.plot(history_plot, color='b', linewidth=2)
    plt.title(f'WartOA Convergence for {function_name}')
    plt.xlabel('Iteration')
    plt.ylabel('Best Fitness Value')

    # Manually set the Y-axis limits if specified
    if config['ylim']:
        plt.ylim(config['ylim'])

    plt.grid(True)
    plt.show()
```
